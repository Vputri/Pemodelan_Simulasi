```
import pandas as pd
from math import factorial, exp

def probablititas(x, miu = 1):
  poisson= ((miu**x) * exp(-miu)) / factorial(x)
  return poisson

def poison(n, p, miu = 1):
  hasil = (miu/n) * p
  return hasil

miu = 5
n = 15

pj = []
pj.append(probablititas(0, miu))

fj = []
fj.append(sum(pj))

for i in range(1, n):
  pj.append(poison(i, pj[-1], miu))
  fj.append(sum(pj))

d = {'p':pj,'F':fj}
df = pd.DataFrame(d)
df
```

```
def random(a, m, c, x0, n):
	x = []
	x.append(x0)
	for i in range(n):
		rand = (a * x[i] + c) % m
		x.append(rand)
	return x

a = 7
m = 99
c = 6
x0 = 3
n = 15

xi = random(a, m, c, x0, n)
U = [(i/m) for i in xi]
```

```
hasil = []

for i in range(len(U)):
	for j in range(len(fj)):
		if U[i] < fj[j]:
			hasil.append(j)
			break
```

```
d = {'x[i]':xi,'U[i]':U, 'p[i]':hasil}
df = pd.DataFrame(d)
df
```

```
xbar = []
for i in range(1, len(hasil)):
  if i == 1:
    result = hasil[1]
  elif i:
    result = result + ((hasil[i]-result)/i)
  xbar.append(result)
```

```
d = {'xbar':xbar}
df = pd.DataFrame(d)
```

```
S = []
for i in range(1,len(xbar)):
  if i == 1:
    result = 18
  else:
    result = (1-(1/i)) * result + i *((xbar[i]-xbar[i-1])**2)
  S.append(result)
```

```
import math
s = []
for i in S:
  s.append(math.sqrt(i))
```

```
hasil = []
for i in range(1, len(s)):
  hasil.append(s[i]/i)
```

```
d = {'s/root_k':hasil}
df2 = pd.DataFrame(d)
new_row = pd.DataFrame({'s/root_k':0}, index =[0])
df2 = pd.concat([new_row, df2]).reset_index(drop = True)
```

```
d = {'s2':S,'s':s}
df1 = pd.DataFrame(d)
new_row = pd.DataFrame({'s2':0, 's':0}, index =[0]) 
df1 = pd.concat([new_row, df1]).reset_index(drop = True) 
result = pd.concat([df, df1, df2], axis=1).reindex(df1.index)
result = result.dropna()
result
```
