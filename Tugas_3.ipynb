# Multiplicative Congruential Method
  
Multiplicative Congruential Method (Lehmer Method) is a type of linear congruential generator for generating pseudorandom numbers in a specific range. This method can be defined as: 

image.png

```
print("Multiplicative Congruential Generator")
print("=====================================\n")

Xo = int(input("Masukkan Nilai Xo      : "))
a = int(input("Masukkan Nilai a       : "))
m = int(input("Masukkan Nilai m       : "))
jmlRandom = int(input("Masukkan Jumlah Random : "))
Nums = [0] * (jmlRandom) 
Nums[0] = Xo 

print("\nJadi, dengan nilai Xo =", Xo,"Nilai a =", a,"dan Nilai m =", m)
print("maka Hasil Random adalah :")

for i in range(1, jmlRandom): 
    Nums[i] = (Nums[i - 1] * a) % m 
    print(Nums[i], end = " ") 
```

# Mixed Congruential Method

```
print("Mixed Congruential Generator")
print("=====================================\n")

Xo = int(input("Masukkan Nilai Xo      : "))
a = int(input("Masukkan Nilai a       : "))
c = int(input("Masukkan Nilai c       : "))
m = int(input("Masukkan Nilai m       : "))
jmlRandom = int(input("Masukkan Jumlah Random : "))
Nums = [0] * (jmlRandom) 
Nums[0] = Xo 

print("\nJadi, dengan nilai Xo =", Xo,"Nilai a =", a,"Nilai c =", c,"dan Nilai m =", m)
print("\nHasil Random adalah :")

for i in range(1, jmlRandom): 
    Nums[i] = ((Nums[i - 1] * a) + c) % m 
    print(Nums[i], end = " ") 
```

```
def f(x) :
    return x

def integral(f, n):
    h = 1 / float(n)
    hasil = 0.5 * h * (f(0) + f(1))
    for i in range(1, int(n)):
        hasil = hasil + h * f(i * h)
    return hasil
 
print("Integral dari X dengan batas 0 dan 1 dan n = 10 adalah",integral(f, 10))
print("Integral dari X dengan batas 0 dan 1 dan n = 100 adalah",integral(f, 100))
print("Integral dari X dengan batas 0 dan 1 dan n = 1000 adalah",integral(f, 1000))
```

**2. Integral dengan Batas a dan b (Riemann Integral)**

Sebuah integral yang didefinisikan dalam sebuah interval [a, b] memberikan nilai dari luas yang tertutup antara fungsi f (x) dan sumbu x dalam sebuah interval [a, b], selama fungsinya kontinu. Untuk menghitung luas yang tertutup di bawah kurva adalah dengan membagi luas tersebut menjadi persegi panjang yang sama dan menjumlahkan luas setiap persegi panjang.

```
a = int(input("Masukkan Batas Awal     : "))
b = int(input("Masukkan Batas Akhir    : "))

def f(x) :
    return 2*x

def integer(f, n):
    lebar = (float(b)-float(a)) /(float(n))
    hasil = 0
    for i in range(n):
        tinggi = f(a + i*lebar)
        luas = lebar * tinggi
        hasil += luas
    return hasil

 
print("Integral dari X dengan batas",a,"dan",b,"dan n = 10 adalah",integer(f,10))
print("Integral dari X dengan batas",a,"dan",b,"dan n = 100 adalah",integer(f,100))
print("Integral dari X dengan batas",a,"dan",b,"dan n = 1000 adalah",integer(f,1000))
```
