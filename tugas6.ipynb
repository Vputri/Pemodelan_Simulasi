```
def random(a, m, c, x0, n):
	x = []
	x.append(x0)
	for i in range(n):
		rand = (a * x[i] + c) % m
		x.append(rand)
	return x

a = 2
m = 33
c = 8
x0 = 7
n = 18
xi = random(a, m, c, x0, n)
U = [(i/m) for i in xi]
```

```
hasil = []

for i in range(len(U)):
	for j in range(len(fj)):
		if U[i] < fj[j]:
			hasil.append(j)
			break

print(hasil)
```

```
d = {'x[i]':xi,'U[i]':U, 'p[i]':hasil}
df = pd.DataFrame(d)
df
```

```
xbar = []
for i in range(1, len(hasil)):
  if i == 1:
    result = hasil[1]
  else:
    result = result + ((hasil[i]-result)/i)
  xbar.append(result)
```

```
S = []
for i in range(len(xbar)):
  if i == 0:
    result = 0
  elif  i == 1:
    result = 18
  else:
    result = (1-(1/i)) * result + i *((xbar[i]-xbar[i-1])**2)
  S.append(result)
```

```
import math

s = []
for i in S:
  s.append(math.sqrt(i))
```

```
hasil = []
for i in range(1, len(s)):
  hasil.append(s[i]/i)
```
